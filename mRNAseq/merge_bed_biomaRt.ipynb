```python
import os
import pandas as pd
```

```python
bed = pd.read_csv(r'E:\work\mRNA-seq\annotation\GRCh37.bed', sep='\t')
biomart = pd.read_csv(r'E:\work\mRNA-seq\annotation\BiomaRt.GRCh37.p13.anno.gz.txt.pared.tsv', sep='\t')

bed.head()
```

```python
biomart.head()
```

```python
bed["Gene stable ID"] = bed["ensembl id"].str.split(".").str[0]

bed.head()
```

```python
bed.shape
```

```python
biomart.shape
```

```python
merged_df = pd.merge(bed, biomart, left_on='Gene stable ID', right_on='Gene stable ID', how='left')

merged_df.head()
```

```python
merged_df = merged_df[['HGNC symbol','gene_name',"ensembl id", 'gene_type', 'gene_name', 'RefSeq ncRNA ID', 'description']]

merged_df.head()
```

```python
merged_df.to_csv(r"E:\work\mRNA-seq\annotation\annotation.tsv", sep="\t", index=False)
```
