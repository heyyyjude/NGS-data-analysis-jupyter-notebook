```python
import pandas as pd
```

```python
cpg_stat = pd.read_csv("merged/insulin-sensitive/Raw_Data/First_CpG_site_stats.txt.out", sep="\t", low_memory=False)
cpg_stat.head()
```

```python
biomart = pd.read_csv('merged/methylation-grch37-p13-mart_export.txt', sep='\t')
biomart.head()
```

```python
df = pd.merge(cpg_stat, biomart, left_on = "Gene", right_on="Gene name", how = 'inner')
df.head()
```

```python
df.to_csv("Second-insensitive-join-two-tsv-files.tsv", sep='\t', index=False)
```

```python
#df_pval = df[df["after.vs.before.pvalue"] <= 0.05]
df_pval = df
df_pval.head()
```

```python
no_pval = df

def get_meth_pos_by_gene_start(row):   
    return int(row["Loc"]) - int(row["Gene start (bp)"])


no_pval["meth.pos.by.gene.start"] = no_pval.apply(get_meth_pos_by_gene_start, axis=1)
no_pval.to_csv('no-pval-sensitive.tsv', sep="\t", index=False)
```

```python
test = df_pval

def get_meth_pos_by_gene_start(row):   
    return int(row["Loc"]) - int(row["Gene start (bp)"])


test["meth-pos-by-gene-start"] = test.apply(get_meth_pos_by_gene_start, axis=1)
test.head()
```
